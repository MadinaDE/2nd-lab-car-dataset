```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data_path = 'cardata.csv'
df = pd.read_csv(data_path)

headers = ["Car_Name","Year","Selling_Price","Present_Prices","Kms_Driven",
           "Fuel_Type","Seller_Type","Transmission","Owner"]

df.columns = headers
df.head()
```

```python
#group the data by the 'Car_Name' column and calculate the average 'Selling_Price' for each car

grouped_df = df.groupby('Car_Name')['Selling_Price'].mean()
```

```python
#bar plot
grouped_df.plot(kind='bar',figsize=(16,7))
plt.title('Average Selling Price by Car Name')
plt.xlabel('Car Name')
plt.ylabel('Average Selling Price')
plt.show()
```

```python
#box plot
#notches can be added to visualise confidence interval around the median
plt.figure(figsize=(20,6))
sns.boxplot(x='Car_Name', y='Selling_Price', data=df, widths=0.5, notch=True)
plt.title('Selling Price Distribution by Car Name')
plt.xlabel('Car Name')
plt.ylabel('Selling Price')
#line rotates 45 degrees and aligns to the right, it helps prevent overlapping labels
plt.xticks(rotation=45, ha='right', fontsize=9)
plt.show()
```

```python
#Swarm Plot

plt.figure(figsize=(20,6))
sns.violinplot(x='Car_Name', y='Selling_Price', data=df)
plt.xticks(rotation=45, ha='right', fontsize=9)
plt.title('Selling Price Distribution by Car Name')
plt.show()
```

```python
# Group the data by 'Car_Name' and 'Fuel_Type' and 'Kms_Driven'

grouped_df = df.groupby(['Car_Name','Fuel_Type'])['Kms_Driven'].mean()
                        
#Visualizing the average mileage by Car Name and Fuel Type using bar plot
                        
grouped_df.unstack().plot(kind='bar', figsize=(16,7))
plt.title('Average Mikleage by Car Name and Fuel Type')
plt.xlabel('Car Name')
plt.ylabel('Average Mileage')
plt.show()
```

```python
#A line plot useful for visualisation trends over time, groupby by 'Car Name' and 'Year' then calcualte the average Selling Price

grouped_df=df.groupby(['Car_Name', 'Year'])['Selling_Price'].mean().reset_index()

#line plot

plt.figure(figsize=(10,6)) 
for Car_Name in grouped_df['Car_Name'].unique():
    subset = grouped_df[grouped_df['Car_Name']==Car_Name]
plt.plot(subset['Year'],subset['Selling_Price'], label=Car_Name)

plt.xlabel('Year')
plt.ylabel('Average Selling Price')
plt.title('Average Selling PricenTrend by Car Name')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# A scatter plot is useful for visualizing the relationship between 2 numerical variables, here the plot the relationship between 
# mileage 'Kms_Driven and'Selling_Price'

plt.figure(figsize=(10,6))
sns.scatterplot(x='Kms_Driven', y='Selling_Price', hue='Fuel_Type', data=df)
plt.xlabel('Kms_Driven')
plt.ylabel('Selling_Price')
plt.title('Selling Price vs. Kms Driven by Fuel Type')
plt.show()
```

```python
#Histogram useful for visualizing the distribution of a numerical variable, here the plot the distribution of prices for different cars

plt.figure(figsize=(12,6))
for Car_Name in df['Car_Name'].unique():plt.hist(df[df['Car_Name']==Car_Name]['Selling_Price'], bins=15, alpha=1, label=Car_Name)
plt.xlabel('Selling_Price')
plt.ylabel('Frequency')
plt.title('Selling Price distribution by Car Name')
plt.legend()
plt.show()
```
